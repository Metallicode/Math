```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer_classification.csv")
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
df.head()
```

##Preprocessing

```python
y = df["benign_0__mal_1"]
```

```python
X = df.drop("benign_0__mal_1", axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##Model Creation

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(30, activation='relu'))

model.add(Dense(15, activation='relu'))

model.add(Dense(1, activation='sigmoid')) ##Sigmoid for binary classification
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

##Using EarlyStopping to stop epochs

```python
callback = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)
```

```python
model.fit(x=X_train_scaled, y=y_train, validation_data=(X_test_scaled, y_test),
          epochs=600, callbacks=[callback])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

##Using Dropout Layer

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5)) ##Drop 50% of nodes
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5)) ##Drop 50% of nodes
model.add(Dense(1, activation='sigmoid'))
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=X_train_scaled, y=y_train, validation_data=(X_test_scaled, y_test),
          epochs=600, callbacks=[callback])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
preds = (model.predict(X_test_scaled) > 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, preds))
```
