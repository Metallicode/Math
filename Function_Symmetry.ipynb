#Function Symmetry

````python
import sympy as sym
````

````python
x = sym.Symbol("x")
````

##Even Function

```
 f(x) = x²
f(-x) = (-x)² = x²
f(-x) = f(x)
```

````python
f_x = x**2
````

````python
sym.plot(f_x)
````

````python
p = sym.Symbol("phi")
p
````

````python
f_p = sym.cos(p)
f_p
````

````python
val = sym.pi/3
print(f_p.subs(p, val))
print(f_p.subs(p, -val))
#when not using pi, we can use evalf on a result to get the actual value.
````

##Odd Function

````python
g_x = x**3
````

````python
sym.plot(g_x)
````

````python
f_p = sym.sin(p)
f_p
````

````python
val = sym.pi/3
f_p.subs(p, val)
````

````python
f_p.subs(p, -val)
````
