#Piecewise Functions

piecewise functions.png

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
xx = np.linspace(-3, 5, 100)
```

```python
p1 = 0 * (xx < 0)
```

```python
p2 = (-2*xx) * ((xx >= 0) & (xx < 3))
```

```python
p3 = xx**3/10 * (xx > 3)
```

```python
y = p1+p2+p3
```

```python
plt.plot(xx, y )
```

##Using Sympy

```python
from sympy.abc import x
```

```python
p_1 = 0
p_2 = -2*x
p_3 = x**3/10
```

```python
f_x = sym.Piecewise((p_1, x<0),(p_2,(x >= 0) & (x < 3)),(p_3,x > 3))
```

```python
f_x
```

```python
sym.plot(f_x, (x, xx[0], xx[-1]))
```

##Find X where f(x) == 0.5

```python
xloc = 0.5

x_s = xx-xloc  #shift the vector with the value we are looking for

#plt.plot(x_s**2)
plt.plot(abs(x_s)) #we can use abs or power
```

```python
index_of_xloc = np.argmin(abs(x_s)) ##look for the min point
```

```python
xx[index_of_xloc]
##notice we don't get 0.5, this is because of the resolution we created the linspace
```
