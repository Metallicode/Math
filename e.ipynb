```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

##estimation of e

```python
n_values = [1, 2, 5, 10, 20, 50]

for n in n_values:
	e = (1+(1/n))**n
	print(f"n: {n:2.0f}, est.e: {e:.5f},  diff to e: {np.exp(1) - e}")
```

##Compare numpy's e with the estimation of e

```python
n = np.arange(1, 1001)
eDiffs = np.exp(1) - (1+1/n)**n

plt.plot(eDiffs)
plt.show()
```

##Power Rules

```python
n = np.arange(-2, 2, 0.01)

y1 = np.exp(n)
y2 = np.exp(n**2)
y3 = np.exp((-n)**2)
y4 = np.exp(-(n**2))
y5 = np.exp(n)**2


plt.plot(n, y1)
plt.plot(n, y2)
plt.plot(n, y3)
plt.plot(n, y4)
plt.plot(n, y5)
plt.show()
```

#sympy.exp and sympy.log

```python
s_beta = sym.var('beta')

s_y = sym.exp(s_beta) - sym.log(s_beta) - np.exp(1)

sym.plot(s_y, (s_beta, -2, 2), title=f'$f(\\beta) = {sym.latex(s_y)}$')
```

##log(exp(x)) Vs exp(log(x))

```python
x = np.linspace(-4, 4, 30)

plt.plot(x, np.exp(x), label='exp(x)')
plt.plot(x, np.log(np.exp(x)), label='log(exp(x))')
plt.plot(x, np.log(x), label='log(x)')
plt.plot(x, np.exp(np.log(x)), label='exp(np.log(x))')
plt.ylim(-2,2)
plt.legend()
plt.show()
```
