##Trigonometric Functions

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
t = np.linspace(-2*np.pi, 2*np.pi, 1000)
````

````python
sin_f = np.sin(t)
cos_f = np.cos(t)
````

````python
plt.plot(t,sin_f)
plt.plot(t,cos_f)
plt.plot(t,sin_f/cos_f)
plt.grid()
plt.ylim(-3,3)
````

````python
plt.plot(t,np.sin(t))
plt.plot(t,np.sin(t**2))
plt.plot(t,np.sin(t)**2)
plt.grid()
plt.ylim(-3,3)
````

##Sin(Cos(θ)) and Cos(Sin(θ))

````python
plt.plot(t,np.sin(t), color="red")
plt.plot(t,np.cos(np.sin(t)), color="green")
plt.plot(t,np.sin(np.cos(t)), color="blue")
plt.grid()
````
