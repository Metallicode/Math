At its core, data visualization is about mapping data values to visual attributes (e.g., position, length, angle, color) in a way that helps people understand, explore, and make sense of the data. Here's a bit more detail on the essential components:

Mapping Data to Visual Attributes:

The most basic form of this is mapping data to positions on axes (like x and y in a scatter plot or x in a histogram). Other attributes include size (as in the size of points in a scatter plot), color (as in the color of bars in a bar chart or points in a scatter plot), shape (as in the shape of points in a scatter plot), etc.
These mappings allow us to perceptually compare data values. For example, position along common scales is very effective for accurately judging differences between values, while color can be effective for understanding categories or relative values.
Aggregation:

This is about summarizing data, usually by grouping it according to some variables (like the 'Country' in a dataset of people around the world), and then computing summary statistics (like mean, sum, count, etc.) for each group.
Examples of visualizations that involve aggregation include bar charts (where the height of the bar represents the count, sum, mean, etc., of some variable for each category), histograms (which count the number of data points within ranges), and box plots (which show median, quartiles, etc., of groups of data).
Transformation:

This can involve various operations to change the scale (e.g., log scales), normalize, or otherwise transform the data. For example, computing the per-capita income from total income and population size, or changing the scale of a plot to logarithmic to handle wide data value ranges.
Layout and Annotation:

Once youâ€™ve made decisions on the above points, you also need to consider how to make the visualization readable and interpretable. This can include adding axis labels, titles, legends, and grid lines, choosing color scales, etc.
The choice of layout (how visual elements are positioned and sized) can also be critical. For example, in a scatter plot, position is used very effectively to represent two continuous variables. In a bar chart, position and length are used to represent categories and a continuous variable, respectively.
Interactivity (in more advanced visualizations):

This includes everything from basic tools to zoom and pan, to more complex interactions like brushing (selecting a region of interest with the mouse) and linking (having multiple plots that respond to user input in coordinated ways).
Narrative and Storytelling:

Beyond the raw presentation of data, many visualizations aim to guide the viewer to certain insights, through the use of annotations, layout, and sequencing (as in a slide show or animation).

##Confidence Interval

[confidence interval (CI) ](https://drive.google.com/file/d/1ypt_YcnIuwB3mch530jQJingpvabiFJ9/view?usp=drive_link)

What It Is:

A Confidence Interval is a range of values we are fairly sure our true value lies in.
Example:

Imagine we are trying to figure out the average height of all people in your city. We can't possibly measure everyone, so we measure 100 people randomly selected, and find the average height of these 100 people is 160 cm.
But that's just the average of these 100 people. How about all the people in the city? We know our sample mean (160 cm) is close but not perfect. The true average height of all people in the city might be a bit less than 160 cm or a bit more.
The Confidence Interval gives us a range, like "from 157 cm to 163 cm". This means we are fairly confident (say, 95% sure) that the true average height of all the people in the city is between 157 cm and 163 cm.

[Calculate STD](https://drive.google.com/file/d/1oBCHqzeC_GAh3ytiuVQYLKN7pyPIb6ny/view?usp=drive_link)


What It Doesn't Mean:

A 95% CI does not mean that the interval contains 95% of the data points, or that there is a 95% chance that a new data point will fall within the interval. It simply means we are 95% confident that the true parameter (like the population mean) is within that interval.

STD Vs CI

### Standard Deviation (STD):

1. **What It Is:**
   - Standard Deviation is a measure of the **spread** or **variability** within a set of data points. It tells us how spread out the numbers in a dataset are around the mean (average) of that dataset.

2. **Usefulness:**
   - It helps to understand how "typical" a data point is within a dataset. A low STD indicates that the data points are close to the mean, whereas a high STD indicates that data points are spread out over a wider range of values.

3. **Example:**
   - Consider two sets of exam scores: Class A: [90, 91, 92, 90, 89] and Class B: [60, 100, 55, 100, 58]. Class A has a lower standard deviation since the scores are closer together, while Class B has a higher standard deviation since the scores are more spread out.

4. **It is a Single Value:**
   - The STD is a single number that tells us about the average distance of each data point from the mean.

5. **When to Use:**
   - Use STD when you want to understand how much individual data points in a dataset vary from the average value.

### Confidence Interval (CI):

1. **What It Is:**
   - A Confidence Interval is a **range of values** that is likely to contain the true value of an unknown parameter (like the mean or proportion of a population). It is calculated from a sample of data and gives an interval, not a single number.

2. **Usefulness:**
   - It helps to give us an estimated range in which we can be fairly certain the true population parameter lies. It comes with a confidence level (e.g., 95%) which quantifies our level of confidence that this interval contains the true parameter.

3. **Example:**
   - From a sample of 100 people, we calculate the average height to be 160 cm. A 95% confidence interval might be 157 cm to 163 cm. This means we are 95% confident that the true average height of the entire population lies in this range.

4. **It is a Range:**
   - The CI gives us a lower and upper value, forming a range where the true parameter is likely to be found.

5. **When to Use:**
   - Use CI when you want to estimate a population parameter (like a population mean) and want to indicate the reliability of that estimate.

### Comparison:

- **STD is about Data Spread:** It tells us how spread out individual data points are around the mean within a single dataset.
- **CI is about Estimation Range:** It tells us in which range we can expect a population parameter (like the population mean) to lie based on a sample of data.
- **STD is Always One Value:** It's a single number, like "5 cm."
- **CI is Always a Range of Values:** It's an interval, like "from 157 cm to 163 cm."
- **Purpose:** STD is used to understand the variability within a single dataset, whereas CI is used to make an inference about a population parameter based on a sample of data.



##Quartiles  [wiki](https://en.wikipedia.org/wiki/Quartile)

Quartiles are a set of values that divide a dataset into four equal parts, or quarters. They are a type of quantile, which are values that divide your data into equal intervals. Here's a simple breakdown of quartiles:

First Quartile (Q1), also known as the lower quartile:

This is the middle number between the smallest number in the data set and the median of the data set. It separates the lowest 25% of the data from the rest. In other words, 25% of the data points are below Q1.
Second Quartile (Q2), also known as the median:

This is the middle value of the data set, separating the data into two halves. Half the scores are greater than or equal to this value and half are less. In terms of percentages, 50% of the data points are below the median, and 50% of the data points are above the median.
Third Quartile (Q3), also known as the upper quartile:

This is the middle value between the median and the highest value of the data set. It separates the lowest 75% of the data from the highest 25%. In other words, 75% of the data points are below Q3.
Interquartile Range (IQR):

This is the range between Q1 and Q3, and it contains the middle 50% of the data points. It is calculated as Q3 - Q1.
The IQR is a measure of variability, and it is resistant to outliers.
Minimum and Maximum (not technically quartiles, but they are often shown on box plots):

The minimum is the smallest data point excluding outliers, and it marks the beginning of the range of the data.
The maximum is the largest data point excluding outliers, and it marks the end of the range of the data.


Quartiles are useful for understanding the distribution of data, spotting outliers, and describing the dataâ€™s spread. They are often visualized using a box plot (or box-and-whisker plot), which shows the median, quartiles, and possible outliers in the data, thereby giving a snapshot of the data distribution.

```python
#sns.boxplot(data=df, x="embarked", y="age")

#transform rotate
sns.boxplot(data=df, y="embarked", x="age")
```

Consider the following dataset:
2,7,8,15,20,21,21,22,23,30

Q1 (First Quartile) = 8.25

Q3 (Third Quartile) = 21.75

IQR = Q3 - Q1 = 21.75 - 8.25 = 13.5

Lower Whisker = Q1 - 1.5 * IQR = 8.25 - 1.5 * 13.5 = -12.5, but the smallest data value is 2, so the lower whisker extends to 2.

Upper Whisker = Q3 + 1.5 * IQR = 21.75 + 1.5 * 13.5 = 42.25, and the largest data value below this is 30, so the upper whisker extends to 30.

The boxplot is a graphical representation of the distribution of a dataset, and it can tell you about the variability of the data and potential outliers. It's a very useful and common way to visualize data distributions.







**Interquartile Range (IQR):**

The IQR represents the middle 50% of the data. It is calculated as the difference between the third quartile (Q3) and the first quartile (Q1):
IQR=Q3âˆ’Q1.
In simple terms, the IQR tells us the range within which the middle 50% of our data values fall. It is a measure of how spread out these values are.
Lower Whisker:

**The lower whisker** of the boxplot extends from the first quartile (Q1) down to the smallest data value that is still within 1.5 times the IQR from Q1.
In simple terms, the lower whisker shows us the range of the lower portion of our data. It tells us where the bulk of the lower values lie, and how spread out they are.
Any data point that falls below the end of the lower whisker is usually considered an outlier, because it is unusually low compared to the rest of the data.
Upper Whisker:

**The upper whisker** of the boxplot extends from the third quartile (Q3) up to the largest data value that is still within 1.5 times the IQR from Q3.
In simple terms, the upper whisker shows us the range of the upper portion of our data. It tells us where the bulk of the higher values lie, and how spread out they are.
Any data point that falls above the end of the upper whisker is usually considered an outlier, because it is unusually high compared to the rest of the data.
Why 1.5 times the IQR?

**The 1.5 is a conventional choice** that
makes the whiskers extend to roughly the maximum and minimum values, unless there are outliers. It is a standard used to formally identify potential outliers. If the whiskers were based on the full range of the data, the outliers would not be distinguishable.

##Heat Map

```python
# Example dataset
data = {
    'Region': ['North', 'South', 'West', 'North', 'South', 'West', 'North', 'South', 'West'],
    'Month': ['January', 'January', 'January', 'February', 'February', 'February', 'March', 'March', 'March'],
    'Product A': [120, 150, 90, 130, 120, 95, 130, 160, 85],
    'Product B': [80, 70, 100, 60, 85, 110, 50, 60, 105],
    'Product C': [200, 210, 190, 220, 230, 210, 240, 235, 200]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Print the dataset
print(df)
```

```python

# Pivot the data - here we are setting Region as index, Month as columns, and the values are the mean of Product A sales
pivot_data = df.pivot_table(values='Product A', index='Region', columns='Month')

# Print the pivoted data
print(pivot_data)

# Draw the heatmap
sns.heatmap(pivot_data, annot=True, cmap='YlGnBu', fmt='g')
```
