#Discontinuous Functions

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
from sympy.abc import x
```

##Jump Discontinuity

```python
p_1 = x/2
p_2 = 1+x**2

f_x = sym.Piecewise((p_1, x<0),(p_2, (x>= 0)&(x<1) ))
```

```python
f_x
```

```python
sym.plot(f_x, (x, -1, 1))
```

##Infinit Discontinuity

```python
sym.plot(sym.sin(x)/sym.cos(x), (x, 0 , sym.pi )) ##tan
```

##Removable Discontinuity

```python
sym.plot( ((x**2)-(2*x)) / ((x**2)-4) , (x, 0 , 5 ))
```

```python
f_x = ((x**2)-(2*x)) / ((x**2)-4)
```

```python
f_x
```

```python
f_x.subs(x,2)
```
