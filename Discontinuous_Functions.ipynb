#Discontinuous Functions

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
from sympy.abc import x
```

##Jump Discontinuity

```python
p_1 = x/2
p_2 = 1+x**2

f_x = sym.Piecewise((p_1, x<0),(p_2, (x>= 0)&(x<1) ))
```

```python
f_x
```

```python
sym.plot(f_x, (x, -1, 1))
```

###Example 2

```python
p_1 = sym.sin(x*sym.pi)
p_2 = 1.5
p_3 = -(x-2)**2

f_x = sym.Piecewise((p_1, x<0),(p_2,sym.Eq(x, 0)), (p_3,x>0))
sym.plot(f_x, (x, -1, 2))
```

```python
f_x
```

##Infinit Discontinuity

```python
sym.plot(sym.sin(x)/sym.cos(x), (x, 0 , sym.pi )) ##tan
```

###Example 2

```python
xx = np.linspace(-2,2,1000)

f_x = 3 / (1-xx**2)
plt.plot(xx,f_x)
plt.ylim(-20,20)
```

```python
sfx = 3 / (1-x**2)
sym.calculus.util.continuous_domain(sfx, x, sym.Interval(-2, 2))
```

```python
sym.singularities(sfx, x)  ##asymptote
```

##Removable Discontinuity

```python
sym.plot( ((x**2)-(2*x)) / ((x**2)-4) , (x, 0 , 5 ))
```

```python
f_x = ((x**2)-(2*x)) / ((x**2)-4)
```

```python
f_x
```

```python
f_x.subs(x,2)
```

###Example 2

```python
xx = np.linspace(-1, 2, 100)
```

```python
y = np.sin(xx*np.pi) + xx**2
y[xx == 0] = np.pi
```

```python
plt.plot(xx, y, "o")
```

##Oscillating Discontinuity

```python
xx = np.linspace(-2, 2, 10000)
```

```python
f_x = np.sin(1 / (xx - 1))
```

```python
plt.xlim(0.5, 1.5)
plt.plot(xx, f_x)
```
