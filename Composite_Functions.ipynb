#Composite Functions

##g(f(x))       Vs       f(g(x))

```python
import sympy as sp
```

```python
x = sp.Symbol("x")
```

```python
x
```

```python
f = 2*x**2 - 4
f
```

```python
g = 7 * x + 3
g
```

```python
g_f = g.subs(x, f)
g_f
```

```python
f_g = f.subs(x, g)
f_g
```

```python
g_f.subs(x, 4)
```

```python
f_g.subs(x, 4)
```

##Example 2

```python
x = sp.Symbol("x")
```

```python
f = sp.sin(x)
f
```

```python
g = sp.sqrt(x)
g
```

```python
h = 2* x**2 + 5
h
```

```python
cf = f.subs(x,g.subs(x, h))
cf
```

```python
cf.subs(x,2)
```

```python
cf.subs(x,2).evalf()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 500)
f_x = 2*x**2 - 4
g_x = 7*np.abs(x) + 3
f_g_x = 2*(7*np.abs(x) + 3)**2 - 4
g_f_x = 7*np.abs(2*x**2 - 4) + 3

plt.figure(figsize=(10,5));
plt.ylim(-10,50)
plt.plot(f_x)
plt.plot(g_x)
plt.plot(f_g_x)
plt.plot(g_f_x)
```
