Vectorizing a function means transforming it so that it can operate on entire arrays of data at once, rather than just on single elements. This is done to improve performance by taking advantage of the low-level vectorized hardware instructions, and to enable cleaner and more readable code. Instead of writing a loop to apply a function to each element of an array one at a time, a vectorized function can process the whole array at once.

```python
#using vectorize will give the result but is more efficient

df["Quality"] = np.vectorize(TipQuality)(df["total_bill"], df["total_bill"])
```

Merge

```python
df1 = pd.DataFrame([["moo",".8"],["blig", ".7"],["plop", ".45"]])
df2 = pd.DataFrame([["moo","45"],["flig", "645"],["plop", "23"]])

df1.index = ['a','b','c']
df2.index = ['d','e','f']

df1.columns = ['X','Y']
df2.columns = ['X','Y']


print(df1)
print()
print(df2)
```

###Inner Join

```python
df1.merge(df2, how='inner', on='X')
```

```python
pd.merge(left=df1, right=df2, how='left', on='X')
```

```python
pd.merge(left=df1, right=df2, how='right', on='X')
```

```python
pd.merge(left=df1, right=df2, how='outer', on='X')
```

```python
##inverse_inner_join


# Perform inner join to find common rows
inner_join = df1.merge(df2, how='inner', on='X')

# Find rows in A that are not in the inner join
A_not_in_B = df1[~df1['X'].isin(inner_join['X'])]

# Find rows in B that are not in the inner join
B_not_in_A = df2[~df2['X'].isin(inner_join['X'])]

# Combine the results (i.e., concatenate A_not_in_B and B_not_in_A)
inverse_inner_join = pd.concat([A_not_in_B, B_not_in_A], ignore_index=True)

# Print the result
print(inverse_inner_join)
```

```python
print(df1['X'].isin(df2['X']))
print(~df1['X'].isin(df2['X']))
```

**left_on, right_on**

The left_on and right_on parameters are extremely useful when the two DataFrames you want to merge don't have columns with the same name to join on, or when you want to join on multiple columns with different names in the two DataFrames.

```python


# Create two DataFrames with columns that we want to join on,
# but the columns have different names in each DataFrame
data1 = {'EmployeeID': [1, 2, 3, 4],
         'EmployeeName': ['Alice', 'Bob', 'Charlie', 'David']}
data2 = {'Emp_ID': [3, 4, 5, 6],
         'Salary': [70000, 80000, 75000, 60000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Merge the DataFrames using the `left_on` and `right_on` parameters
result = pd.merge(df1, df2, left_on='EmployeeID', right_on='Emp_ID', how='inner')

# Print the result
print(result)
```

**pivot_data**

The pivot_table function allows you to specify columns whose unique values will become the new column headers and rows whose unique values will become the new row indices, along with one or more columns whose data will be aggregated using a specified function (such as 'mean', 'sum', 'count', etc.).

```python

# Sample sales data
data = {'Date': ['2023-08-01', '2023-08-01', '2023-08-02', '2023-08-02', '2023-08-03'],
        'Product': ['A', 'B', 'A', 'A', 'B'],
        'Sales': [100, 150, 200, 50, 300]}

df = pd.DataFrame(data)

# The raw data looks like this:
print(df)
```

```python
pivot_table = pd.pivot_table(df, values='Sales', index='Date', columns='Product', aggfunc='sum', fill_value=0)

# The resulting pivot table looks like this:
print(pivot_table)
```

The unique values of the Date column become the index (row labels) of the pivot table, the unique values of the Product column become the columns of the pivot table, and the Sales values are summed wherever they share the same Date and Product. We specified fill_value=0, so any missing values (where a product wasn't sold on a particular date) are filled with zeros.

You can apply multiple aggregation functions at once by passing a list to the aggfunc parameter. For example:

```python
pivot_table = pd.pivot_table(df, values='Sales', index='Date', columns='Product', aggfunc=['sum', 'mean'], fill_value=0)
print(pivot_table)
```
